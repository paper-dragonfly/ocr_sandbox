## Crop and Resize

* Takes raw image and crops everything outside of the data box
* Resizes image so hight is 500 pxls

```python
import cv2
from matplotlib import pyplot as plt
import pytesseract
```

```python
image_id = 5
file_name = f'temp/erg_0{str(image_id)}.jpeg'
original = cv2.imread(file_name)
print(type(original))
print(original.shape)
plt.imshow(original)
```

```python
yt = 200
yb = 525
xl = 5
xr = 360
cropped = original[yt:yb, xl:xr]
print(cropped.shape)
plt.imshow(cropped)
```

```python
#Resize
hf = 500
wi = cropped.shape[1]
hi = cropped.shape[0]
wf = int(wi * hf/hi)
resized = cv2.resize(cropped, (wf, hf))
print(resized.shape)
plt.imshow(resized)
```

```python
#Save 
cv2.imwrite(f'temp/cr_erg0{str(image_id)}.jpg', resized)
```
