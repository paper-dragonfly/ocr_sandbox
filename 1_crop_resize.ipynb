## Crop and Resize

* Takes raw image and crops down to be screen only
* Resizes image so hight is 500 pxls

```python
import cv2
from matplotlib import pyplot as plt
import pytesseract
```

```python
file_name = 'temp/cr_erg02.jpg'
original = cv2.imread(file_name)
print(original.shape)
plt.imshow(original)
```

```python
yt = 0
yb = 505
xl = 0
xr = 510
cropped = original[yt:yb, xl:xr]
print(cropped.shape)
plt.imshow(cropped)
```

```python
#Resize
hf = 500
wi = cropped.shape[1]
hi = cropped.shape[0]
wf = int(wi * hf/hi)
resized = cv2.resize(cropped, (wf, hf))
print(resized.shape)
plt.imshow(resized)
```

```python
#Save 
cv2.imwrite('temp/cr_erg02.jpg', resized)
```
